```python
import pandas as pd
```

```python
netflix_titles_df = pd.read_csv("titles.csv")
display(netflix_titles_df)
```

```python
netflix_df = netflix_titles_df[['title', 'type', 'imdb_score', 'release_year','age_certification']]
display(netflix_df)
netflix_df.info()
```

```python
netflix_df_group = netflix_df.groupby(by=['imdb_score','type']).count()
display(netflix_df_group)
```

```python
netflix_filmes_df = netflix_df.loc[netflix_df['type']=="MOVIE"]
display(netflix_filmes_df)
netflix_series_df = netflix_df.loc[netflix_df['type']=="SHOW"]
display(netflix_series_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15,15))
netflix_eixo_grafico = plt.subplot()
#primeiro plot (filme)

y_filmes = netflix_filmes_df['imdb_score']
x_filmes = netflix_filmes_df['release_year']
y_series = netflix_series_df['imdb_score']
x_series = netflix_series_df['release_year']
plt.grid(axis="both")

#colors = np.where(netflix_df['type']=="MOVIE",'RED','BLUE')
#plt.scatter(x, y, c=colors, alpha=0.3, label=colors)
plt.scatter(x_filmes,y_filmes,c='blue',label='Filme',alpha=0.8)
plt.scatter(x_series,y_series,c='red',label='Series',alpha=0.8)
plt.legend(loc="upper left")
netflix_eixo_grafico.set_xlabel("Ano de Lançamento")
netflix_eixo_grafico.set_ylabel("Nota IMDB")
netflix_eixo_grafico.set_title("Programas na Netflix Americana - Nota IMDB x Ano de Lançamento original")
#netflix_eixo_grafico.legend(netflix_df['type'])
plt.savefig("grafico_netflix.png", dpi = 150)
plt.show()
```

```python
import openpyxl
excel = "grafico_netflix.xlsx"


wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Gráfico"
wb.save(excel)
grafico = openpyxl.drawing.image.Image('grafico_netflix.png')
ws.add_image(grafico, 'A1')
wb.save(excel)
```
